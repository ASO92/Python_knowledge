## Passign values by reference modifies the global varaible



```python
def change_list(fruits_list):
    
    fruits_list[0] = "Kiwi"
    
    print("Inside the function: ", fruits_list)

change_list(fruits_list)
print()
print("Outside the function: ", fruits_list)
```

### Modifying a complex data type by reference (reasign the values of the list)

**Modifies the actual data outside the function as well**

### Reassigning a complex data type: fruits_list = ["Kiwi"]

**This is reassigning the list inside the function - does not affect the original list**

```python
fruits_list= ["Apple", "Grapes"]
```

```python
change_list(fruits_list)
```

```python
fruits_list
```

### Appending is modyfing, not reassigning

```python
fruits_list = ["Apple", "Grapes", "Mango", "Bananas"]

def change_list(fruits_list):
    
    fruits_list.append("Kiwi")
    
    print("Inside the function: ", fruits_list)
    
change_list(fruits_list)
print()
print("Outside the function: ", fruits_list)
```

### Deleting is modyfing, not reassigning

```python
fruits_list = ["Apple", "Grapes", "Mango", "Bananas"]

def change_list(fruits_list):
    
    fruits_list.remove("Apple")
    
    print("Inside the function: ", fruits_list)
    
change_list(fruits_list)
print()
print("Outside the function: ", fruits_list)
```

### Sorting and poppint are also modyfing, not reassigning



```python
fruits_list = ["Apple", "Grapes", "Mango", "Bananas"]

def sort_items_and_remove_last(my_list):
    my_list.sort()
    my_list.pop()
    
    print("Inside the function", my_list)
```

```python
sort_items_and_remove_last(fruits_list)
```

```python
fruits_list
```

### Dictionary modification. Also modifies the global variable adding new item and deleting.

```python
car_speeds_dictionary = {
    "Camry": 120, 
    "Accord":130, 
    "Cayenne": 250,
    "Mini Cooper": 100
}
```

```python
def change_dictionary(car_speeds_dictionary):
    
    car_speeds_dictionary["Mustang"] = 200
    
    print("Inside the fucntion: ", car_speeds_dictionary)
```

```python
change_dictionary(car_speeds_dictionary)
```
