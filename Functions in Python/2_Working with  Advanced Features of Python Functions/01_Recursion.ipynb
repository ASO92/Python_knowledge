## Recursion

**Invoking a function from within the function itself**

```python
def hello(name):
    
    print("Hello", name)
    hello(name)
```

## Infinite recursion

**Can cause you to run out of memory and crash**

```python
hello("Ron") #Infinite loop. RecursionError: maximum recursion depth exceeded while calling a Python object
```

### Default Recursion limit

```python
import sys
sys.getrecursionlimit()
```

### La cantidad de recursin que tiene, no tiene porqu ser 3000 exactos, depende del input buffering and so on.

**Se puede ver que en el siguiente bucle no se llega a 3000**

```python
def increment(num):
    
    print(num, end = " ")
    
    increment(num + 1)
    
    return
```

```python
increment(1)
```

### Set recursion limit

**The final number is not necessary close to 100. This is because there is some buffering going on at the output and it's possible that Python just didn't print the last few numbers before the error was encountered**

```python
sys.setrecursionlimit(100)
```

```python
increment(1)
```

## Conditions

**The first thing that you ought to put when defining a recursion function is the terminating condition**

**Next example the same funcion without or with recursion**

```python
def decrement(num):
    
    while num > 0:
        print(num)
        num = num - 1
```

```python
decrement(10)
```

```python
def decrement_recursion(num):
    
    if num == 0: # Dominating condition
        return # return when num is 0

    print(num)
    decrement(num -1 )
```

```python
decrement_recursion(10)
```

### Decrement Sum explanation

**First go down in the function and after go up**

recursive_sum(5) = 5 + recursive_sum(4)

recursive_sum(4) = 4 + recursive_sum(3)

recursive_sum(3) = 3 + recursive_sum(2)

recursive_sum(2) = 2 + recursive_sum(1)

recursive_sum(1) = 1 + recursive_sum(0)

recursive_sum(0) = 0

**Si deshacemos los cambios de variable**

recursive_sum(0) = 0

recursive_sum(1) = 1 + recursive_sum(0) = 1

recursive_sum(2) = 2 + recursive_sum(1) = 3

recursive_sum(3) = 3 + recursive_sum(2) = 6

recursive_sum(4) = 4 + recursive_sum(3) = 10

recursive_sum(5) = 5 + recursive_sum(4) = 15




```python
def recursive_sum(num):
    
    if num == 0:
        return 0
    
    result = num + recursive_sum(num -1 )
    
    return result
```

```python
sys.setrecursionlimit(1000)
recursive_sum(5)
```

```python
recursive_sum(3)
```

## Recursion - calls

### Funcin factorial

**factorial(3) = 3*2*1**

**factorial(3) = 3 * factorial(2)**






```python
def factorial(number):
    
    if number < 0:
        print("Sorry, factorial does not exist for negative numbers")
        return
    
    if number == 0:
        return 1
    
    return number * factorial(number - 1)
```

```python
factorial(3)
```

### Fibonacci

```python
def fibonacci(number, fib_series):
    
    if number < 2:
        return
    
    l = len(fib_series) #len devuelve el nmero total de elementos. Ej: l1=[0,1,2,3,4] len[l1] = 5
    
    new_number = fib_series[l - 1] + fib_series[l - 2] # indice [l -1] es el ltimo nmero, [l -2] es el penltimo nmero
    
    fib_series.append(new_number)
    
    print("Series so far", fib_series)
    
    fibonacci(number -1, fib_series)
```
